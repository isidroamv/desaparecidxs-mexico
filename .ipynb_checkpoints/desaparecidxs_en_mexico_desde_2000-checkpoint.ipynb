# Desaparecidos  en México por año

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('report_12_01_2018_2.csv')
```

```python
# Count records by year
years = []
values = []
current_year = 2008
while True:
    true_identities = df['fuerocomun_desapfecha'].str.contains(str(current_year))
    identities = df[true_identities]
    if len(identities) > 0:
        years.append(current_year)
        values.append(len(identities))

    if current_year == 2020:
        break
    
    current_year += 1

labels = years
```

```python
# Colors
colors = ['#3366cc','#dc3912','#ff9900','#109618','#990099',
 '#0099c6','#dd4477','#66aa00','#b82e2e','#316395',
 '#3366cc','#994499','#22aa99','#aaaa11','#6633cc',
 '#e67300','#8b0707','#651067','#329262','#5574a6',
 '#3b3eac','#b77322','#16d620','#b91383','#f4359e',
 '#9c5935','#a9c413','#2a778d','#668d1c','#bea413',
 '#0c5922','#743411']
```

```python
# Show Chart
plt.scatter(years, values, values, c=colors, alpha=0.4)
plt.grid(color='#D3D3D3', linestyle='-', linewidth=1)
plt.xticks(years)
plt.title("Desaparecidos por año")
for i, txt in enumerate(values):
    plt.annotate(str(txt), (years[i], values[i]))

plt.show()
```
