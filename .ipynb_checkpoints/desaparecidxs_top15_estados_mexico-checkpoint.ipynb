# Ciudades de México con más desaparecidos

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('report_12_01_2018_2.csv')
```

```python
state_count = df['fuerocomun_desaplocalidad'].value_counts()
top_state_count = state_count.head(n=15)
```

```python
labels = ["Otros"]
values = [0]
for key_val in top_state_count.iteritems():
    labels.append(key_val[0])
    values.append(key_val[1])


for key_val in state_count.iteritems():
    if key_val[0] in labels:
        continue
    values[0] += key_val[1]
```

```python
# Generate Chart
colors = ['#3366cc','#dc3912','#ff9900','#109618','#990099',
 '#0099c6','#dd4477','#66aa00','#b82e2e','#316395',
 '#3366cc','#994499','#22aa99','#aaaa11','#6633cc',
 '#e67300','#8b0707','#651067','#329262','#5574a6',
 '#3b3eac','#b77322','#16d620','#b91383','#f4359e',
 '#9c5935','#a9c413','#2a778d','#668d1c','#bea413',
 '#0c5922','#743411']

plt.pie(values, labels=labels, autopct='%1.1f%%', shadow=True,  labeldistance=1.05)
plt.show()
```
